Converting database into dataframe

```python
!pip install pymysql
import pymysql
import pandas as pd
```

```python
SQL_Query = pd.read_sql_query(
        '''select 
          STATE,
          CASES,
          ACTIVE,
          RECOVERED,
          DEATHS from CoronavirusOutbreak''', connection)

df1 = pd.DataFrame(SQL_Query, columns=['STATE', 'CASES', 'ACTIVE','RECOVERED','DEATHS'])
df1
```

```python
SQL_Query = pd.read_sql_query(
        '''select 
          STATE,
          DOSE1,DOSE2,
          TOTAL_VACCINATION from VACCINE_TRACKER''', connection)

df2 = pd.DataFrame(SQL_Query, columns=['STATE', 'DOSE1', 'DOSE2','TOTAL_VACCINATION'])
df2
```

```python
statewise=pd.pivot_table(df1,values=["ACTIVE","RECOVERED","DEATHS"],index="STATE",aggfunc=max)
statewise["Recovery Rate"]=statewise["RECOVERED"]*100/statewise["ACTIVE"]
statewise["MORTALITY Rate"]=statewise["DEATHS"]*100/statewise["ACTIVE"]
statewise=statewise.sort_values(by="ACTIVE",ascending=False)
statewise.style.background_gradient(cmap="cubehelix")
```

Sorting dataframe

```python
statewise=pd.pivot_table(df2,values=["DOSE1","DOSE2","TOTAL_VACCINATION"],index="STATE",aggfunc=max)
statewise=statewise.sort_values(by="DOSE1",ascending=False)
statewise.style.background_gradient(cmap="cubehelix")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
top_10_deaths=df1.groupby(by='STATE',as_index=True).max()[['DEATHS','CASES']].sort_values(by=['DEATHS'],ascending=False).reset_index()
fig=plt.figure(figsize=(25,5))
plt.title("TOP 10 STATES WITH MOST DEATHS", size=25)
ax=sns.barplot(data=top_10_deaths.iloc[:12],y="DEATHS",x="STATE",linewidth=2,edgecolor='black')
plt.xlabel("states")
plt.ylabel("total deaths cases")
plt.show()
top_10_deaths=df1.groupby(by='STATE',as_index=True).max()[['ACTIVE','CASES']].sort_values(by=['ACTIVE'],ascending=False).reset_index()
fig=plt.figure(figsize=(25,5))
plt.title("TOP 10 STATES WITH MOST ACTIVE CASES", size=25)
ax=sns.barplot(data=top_10_deaths.iloc[:12],y="ACTIVE",x="STATE",linewidth=2,edgecolor='black')
plt.xlabel("states")
plt.ylabel("total active cases")
plt.show()
top_10_deaths=df1.groupby(by='STATE',as_index=True).max()[['RECOVERED','CASES']].sort_values(by=['RECOVERED'],ascending=False).reset_index()
fig=plt.figure(figsize=(25,5))
plt.title("TOP 10 STATES WITH MOST RECOVERED CASES", size=25)
ax=sns.barplot(data=top_10_deaths.iloc[:12],y="RECOVERED",x="STATE",linewidth=2,edgecolor='black')
plt.xlabel("states")
plt.ylabel("total recovered cases")
plt.show()
```

```python
df3=pd.concat( [df1, df2], axis = 1, ignore_index=False)
df3
```

```python
active=df3["ACTIVE"].mean()
recovered=df3["RECOVERED"].mean()
deaths=df3["DEATHS"].mean()
cases=df3["CASES"].mean()
dose1=df3["DOSE1"].mean()
dose2=df3["DOSE2"].mean()
total_vac=df3["TOTAL_VACCINATION"].mean()
px.pie(names=["ACTIVE","RECOVERED","DEATHS","CASES","DOSE1","DOSE2","TOTAL_VACCINATION"],values=[active,recovered,deaths,cases,dose1,dose2,total_vac],title="COVID 19 UPDATES")
```
